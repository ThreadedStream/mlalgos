```python
import django
import os
import pandas as pd
import numpy as np
import threading
os.environ["DJANGO_SETTINGS_MODULE"] = 'mlproject.settings'
django.setup()

from bookapp.models import *
```

```python
PATH = '../datasets/books.csv'
df = pd.read_csv(PATH, error_bad_lines=False, usecols=['title', 'authors'])
df.head(20)
```

```python
def generate_prices(n):
    for i in range(n):
        yield np.random.randint(30, 150)
```

```python
already_done = False
def transform(df):
    prices = list(generate_prices(df.shape[0]))
    titles = list(df.loc[:, 'title'])
    authors = list(df.loc[:, 'authors'])
    for i in range(df.shape[0]):
        Book.objects.create(title=titles[i], author=authors, price=prices[i])
    already_done = True
    
```

```python
tr = threading.Thread(target=transform, args=(df,))
if not already_done:
    tr.start()
```
